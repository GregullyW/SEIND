```python
import load_csv_parallel

tables = load_csv_parallel.load_tables_parallel("../data/SISU")
```

```python
def column_to_text(table_name, column):
    return f"This column is named {column} from table {table_name}."
```

```python
def extract_columns(tables, dataset_name):
    columns = []

    for table_name, df in tables.items():
        for col in df.columns:
            try:
                text = column_to_text(
                    table_name,
                    col
                )

                columns.append({
                    "dataset": dataset_name,
                    "table": table_name,
                    "column": col,
                    "text": text
                })

            except Exception as e:
                print(f"Erro {table_name}.{col}: {e}")

    return columns
```

```python
dataset_name = "SISU"
num_tables = 2
columns_data = extract_columns(tables, dataset_name)
```
